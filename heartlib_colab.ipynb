# heartlib Colab Runner

这个 notebook 会：
1. 安装 uv
2. 初始化项目依赖
3. clone `heartlib`
4. 下载 checkpoint
5. 运行一次短音频生成


```python
import os
import subprocess

PROJECT_DIR = '/content/heartlib-colab'
REPO_URL = 'https://github.com/topaihub/heartlib-colab.git'

if not os.path.isdir(PROJECT_DIR):
    subprocess.run(['git', 'clone', REPO_URL, PROJECT_DIR], check=True)
else:
    subprocess.run(['git', '-C', PROJECT_DIR, 'fetch', 'origin'], check=True)
    subprocess.run(['git', '-C', PROJECT_DIR, 'checkout', 'main'], check=True)
    subprocess.run(['git', '-C', PROJECT_DIR, 'reset', '--hard', 'origin/main'], check=True)
    subprocess.run(['git', '-C', PROJECT_DIR, 'clean', '-fd'], check=True)

print('Project dir:', PROJECT_DIR)
```
