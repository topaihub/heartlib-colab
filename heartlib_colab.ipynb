# heartlib Colab Runner

这个 notebook 会：
1. 强制同步最新 `heartlib-colab` 代码
2. 安装 uv 和项目依赖
3. clone 官方 `heartlib`
4. 下载 checkpoint
5. 运行一次短音频生成


```python
import os
import shutil
import subprocess

PROJECT_DIR = '/content/heartlib-colab'
REPO_URL = 'https://github.com/topaihub/heartlib-colab.git'

if os.path.isdir(PROJECT_DIR):
    shutil.rmtree(PROJECT_DIR)

subprocess.run(['git', 'clone', REPO_URL, PROJECT_DIR], check=True)
subprocess.run(['git', '-C', PROJECT_DIR, 'log', '--oneline', '-n', '1'], check=True)

print('Project dir:', PROJECT_DIR)
```
